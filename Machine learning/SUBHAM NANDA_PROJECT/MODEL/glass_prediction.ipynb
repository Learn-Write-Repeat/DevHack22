The diamonds shaped above the box plot gives us an idea about that there are some outliers present in the data. 

So we must try to eliminate the outliers so as to find maximum accuracy. 

```python
# Selecting the features from the dataset   

features = glass_data.columns[:-1].tolist()
from collections import Counter
```

```python
# outliers detection

# this algorithm is written so as to find the outliers in the data. 

def find(glass_data):

    outlier_indices = []
    
    # iterate over features(columns)
    for col in glass_data.columns.tolist():
        # 1st quartile
        Q1 = np.percentile(glass_data[col], 25)
        
        # 3rd  quartile
        Q3 = np.percentile(glass_data[col],75)
        
        # Interquartile rrange (IQR)
        IQR = Q3 - Q1
        
        # outlier step
        outlier_step = 1.5 * IQR
        
        # this determines a list of indices of outliers for feature col
        outlier_list_col = glass_data[(glass_data[col] < Q1 - outlier_step) | (glass_data[col] > Q3 + outlier_step )].index
        
        # append the found outlier indices for col to the list of outlier indices 
        outlier_indices.extend(outlier_list_col)
        
    #  To select the observations containing more than 2 outliers
    outlier_indices = Counter(outlier_indices)        
    multiple_outliers = list( k for k, v in outlier_indices.items() if v > 2 )
    
    return multiple_outliers   

print('This set of data contains %d observations with more than 2 outliers' %(len(find(glass_data[features])))) 
```

### MODEL DEVELOPMENT

```python
# Splitting the data into two variables 

X = glass_data.drop("Type", axis = 1).values
y = glass_data["Type"].values.reshape(-1,1)
```

```python
#  Splitting the data into train and test with 70% for training the model and 30% for testing the model. 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Check whether the train and test have instances of all the glass types

print(np.unique(y_train))
print(np.unique(y_test))
```

```python
# to find no. of rows and columns in each type of data

print("Shape of X_train: ",X_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test",y_test.shape)
```

##### Importing the required libraries for Model Creation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
```

```python
# LOGISTIC REGRESSION 

logReg = LogisticRegression(max_iter = 5000)
logReg.fit(X_train, y_train)
logReg.score(X_train, y_train)
logReg.score(X_test, y_test)
y_pred_LR = logReg.predict(X_test)
```

```python
print('Accuracy of Logistic Regression Model:' ,metrics.accuracy_score(y_test,y_pred_LR))
```

```python
# RANDOM FOREST CLASSIFIER

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc.score(X_train, y_train)
rfc.score(X_test, y_test)
y_pred_rfc = rfc.predict(X_test)
```

```python
print('Accuracy of Random Forest Classifier Model:' ,metrics.accuracy_score(y_test,y_pred_rfc))
```

```python
# DECISION TREE CLASSIFIER

DecTree = DecisionTreeClassifier()
DecTree.fit(X_train, y_train)
y_pred_DT = DecTree.predict(X_test)
```

```python
print('Accuracy of DecisionTreeClassifier Model:' ,metrics.accuracy_score(y_test,y_pred_DT))
```

```python
# SUPPORT VECTOR CLASSIFIER

svm = SVC()
svm.fit(X_train,y_train)
y_pred_svm = svm.predict(X_test)
```

```python
print('Accuracy of Support Vector Model:' ,metrics.accuracy_score(y_test,y_pred_svm))
```

```python
# GAUSSIAN NAIVE BAYES CLASSIFIER

gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb.score(X_train, y_train)
gnb.score(X_test, y_test)
y_pred_gnb = gnb.predict(X_test)
```

```python
print('Accuracy of Gaussian Naive Bayes Classifier Model:' ,metrics.accuracy_score(y_test,y_pred_gnb))
```

```python
# GRADIENT BOOSTING CLASSIFIER 

gfc=GradientBoostingClassifier(n_estimators= 1000, max_leaf_nodes= 4, max_depth=None,random_state= 2,min_samples_split= 5)
gfc.fit(X_train,y_train)
gfc.score(X_train, y_train)
gfc.score(X_test, y_test)
y_pred_gfc = gfc.predict(X_test)
```

```python
print('Accuracy of Gradient Boosting Classifier Model:' ,metrics.accuracy_score(y_test,y_pred_gfc))
```

## Comparing the accuracy of models

```python
pd.DataFrame([['Logistic Regression',metrics.accuracy_score(y_test,y_pred_LR)],['Gradient Boost',metrics.accuracy_score(y_test,y_pred_gfc)], ['Gaussian Naive Bayes',metrics.accuracy_score(y_test,y_pred_gnb)],
                             ['Decision Tree',metrics.accuracy_score(y_test,y_pred_DT)],['Random Forest',metrics.accuracy_score(y_test,y_pred_rfc)]],
                                 columns=['Model','Accuracy'])
```

#### From the above table we found that the Gradient Boosting Algorithm works best in our dataset

###### Hence we will have that model and make final prediction and find the appropriate accuracy of our model

### FINAL MODEL CREATION 

```python
# creating the final model and saving it. 

import pickle
filename = 'final_glass_prediction_model.pkl'
pickle.dump(gfc, open(filename, 'wb'))
final_model = pickle.load(open(filename, 'rb'))
result = final_model.score(X_test, y_test)
print(result)
```
